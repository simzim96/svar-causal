```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates



data = pd.read_csv('../data/transformed_data/all_time_series.csv', index_col=0)
df = data
df. rename(columns={'art':'Digital Art Market UTD'}, inplace=True)
df. rename(columns={'hype':'Collectible Market Index UTD'}, inplace=True)
df. rename(columns={'gf_index':'Greed and Fear Index CNN'}, inplace=True)
df. rename(columns={'spy':'SPY ETF'}, inplace=True)
```

```python
# sotheby's
# 6 February 2023 CryptoPunks #5066 857 ETH 1,383,464 MIo USD
# 6 June 2023 Fidenza 545 $1,178,850.00
# 24 Juli Vera Molner Sotheby
# 
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(12, 10))
plt.plot(df.index, df['Digital Art Market UTD'], label='Digital Art Market UTD', color='cyan')
plt.plot(df.index, df['Collectible Market Index UTD'], label='Collectible Market UTD', color='magenta')
plt.plot(df.index, df['Greed and Fear Index CNN'], label='Greed and Fear CNN', color='yellow')
plt.plot(df.index, df['SPY ETF'], label='SPY ETF', color='green')

plt.title('Art and Financial Market Indicators')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.axvline(x='2023-06-15', color='white', linestyle='--', label='3AC Sotheby\'s Sale')
plt.axvline(x='2023-02-06', color='white', linestyle='--', label='1.3 Mio Larva Labs Sale')
plt.axvline(x='2023-07-24', color='white', linestyle='--', label='Sotheby\'s Vera Molner Sale')
plt.axvline(x='2023-06-06', color='white', linestyle='--', label='1.17 Mio Fidenza Sale')
# show legend avxline


# Setting the major and minor grid lines to monthly
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_minor_locator(mdates.MonthLocator())
# make all text bigger
plt.rc('font', size=15)
plt.grid(True, which='major', color='gray', linestyle='-')
plt.grid(True, which='minor', color='gray', linestyle='--')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust the layout
plt.show()

```
